```python
import pandas as pd
gdp = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita",attrs={'class':'wikitable sortable'},index_col=1)[0]
gdp
```

```python
hap = pd.read_html("https://en.wikipedia.org/wiki/World_Happiness_Report",attrs={'class':'wikitable sortable'},index_col=1)[0]
hap
```

```python
exp = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy",attrs={'class':'wikitable nowrap sortable mw-datatable'},index_col=0)[0]
exp
```

```python
sov = pd.read_html("https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_birth_rate",attrs={'class':'sortable wikitable'},index_col=0)[0]
sov
```

```python
birthrate = sov.iloc[:,[8]]
birthrate
```

```python
life = exp.iloc[:,[2]]
life
```

```python
hapscore = hap.iloc[:,[1]]
hapscore
```

```python
gdppc = gdp.iloc[:,[1]]
gdppc
```

```python
df = pd.concat([gdppc, hapscore, life, birthrate], axis=1)
df
```

```python
labels = "GDP (PPP) per capita,Happiness score,Life expectancy,Birth rate"
labels = labels.split(",")
df.columns = labels
df
```

```python
html = df.to_html("countrydata.html")
csv = df.to_csv("countrydata.csv")
json = df.to_json("countrydata.json")
```

```python
dfjson = pd.read_json("countrydata.json")
dfjson
```

```python
import matplotlib.pyplot as plt
plt.scatter(dfjson["GDP (PPP) per capita"], dfjson["Happiness score"])
plt.xlabel("GDP (PPP) per capita")
plt.ylabel("Happiness score")
plt.show()
```
