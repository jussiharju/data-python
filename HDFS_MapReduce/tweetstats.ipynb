```python
%%writefile tweetstats.py

from mrjob.job import MRJob
import mrjob

class TweetStats(MRJob):
    
    OUTPUT_PROTOCOL = mrjob.protocol.RawValueProtocol
    
    def mapper(self, _, line):
        try:
            date = line.split(',')
            date = date[len(date)-5].split(' ')
            hour = date[1].split(':')[0]
            year = date[0].split('-')[2]
            yield (year, int(hour)), 1
        except IndexError:
            pass
    
    def combiner(self, key, value):
        yield int(key[0]), (key[1], sum(value))
        
    def reducer(self, key, value):
        if key == 2009:
            days = 242
        elif key == 2019:
            days = 327
        elif key % 4 == 0:
            days = 366
        else:
            days = 365

        yearlyTweets = 0
        
        hourly = {}
        for i in range(24):
            hourly[i] = 0
        
        for v in value:
            yearlyTweets += v[1]
            hourly[v[0]] += v[1]
        
        output = str(key)
        for i in range(24):
            output += " " + str(int(hourly[i]/yearlyTweets*100)).zfill(2)
        output += " " + str(round(yearlyTweets/days, 1))
        yield None, output

if __name__ == '__main__':
    TweetStats.run()
```
