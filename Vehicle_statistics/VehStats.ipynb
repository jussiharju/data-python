```python
import pandas as pd
veh = pd.read_csv("Tieliikenne Avoin Data 5.6.csv", sep=";", encoding="latin")
veh.head(5)
```

```python
labels = "vehicleClass;registrationDate;vehicleGroup;usageType;variant;version;deploymentDate;colour;numDoors;chassisType;cabinType;numSeats;weight;vehMaxLoad;regMaxLoad;length;width;height;fuelType;engineCapacity;power(kW);numCylinders;turbo;hybrid;hybridClass;make;model;transmission;numGears;name;brakeType;modelRegNumber;euroFuelType;municipality;co2;mileage(km);serialNumber;lineNumber"
labels = labels.split(";")
veh.columns
veh.columns = labels
veh.head(5)
```

```python
veh = veh.loc[(veh['vehicleClass'] == 'M1') | (veh['vehicleClass'] == 'M1G')]
veh.head(10)
```

```python
veh = veh.loc[:,["make","deploymentDate","mileage(km)","fuelType","co2"]]
veh.head(10)
```

```python
df = veh
veh["deploymentDate"] = veh["deploymentDate"].astype(str)
veh["deploymentDate"] = veh["deploymentDate"].str.slice(0, 4)
veh["deploymentDate"] = pd.to_datetime(veh["deploymentDate"], format='%Y', errors='coerce')
veh.head(10)
```

```python
veh["mileage(km)"] = veh["mileage(km)"].abs()
veh["fuelType"] = pd.to_numeric(veh["fuelType"], errors='coerce')
veh.head(10)
```

```python
veh = veh.dropna()
veh.head(10)
```

```python
import matplotlib.pyplot as plt
counts = veh["make"].value_counts()[0:5]
others = pd.Series([sum(veh["make"].value_counts()[5:])], index=['Others'])
counts = counts.append(others)
```

```python
#veh.loc[veh["mileage(km)"] <= 1000000]["mileage(km)"].hist(bins=1000)
plt.hist(veh.loc[veh["mileage(km)"] <= 1000000]["mileage(km)"], bins=1000)
```

```python
from datetime import datetime
u5 = veh.loc[veh["deploymentDate"] >= datetime(2014,1,1)]

fiveto10 = veh.loc[(veh["deploymentDate"] < datetime(2014,1,1)) & (veh["deploymentDate"] >= datetime(2009,1,1))].loc[:,["deploymentDate"]]

tento15 = veh.loc[(veh["deploymentDate"] < datetime(2009,1,1)) & (veh["deploymentDate"] >= datetime(2004,1,1))].loc[:,["deploymentDate"]]

fifteento20 = veh.loc[(veh["deploymentDate"] < datetime(2004,1,1)) & (veh["deploymentDate"] >= datetime(1999,1,1))].loc[:,["deploymentDate"]]

twentyto25 = veh.loc[(veh["deploymentDate"] < datetime(1999,1,1)) & (veh["deploymentDate"] >= datetime(1994,1,1))].loc[:,["deploymentDate"]]
```

```python
gas = veh.loc[veh["fuelType"] == 1.0]["co2"].tolist()
die = veh.loc[veh["fuelType"] == 2.0]["co2"].tolist()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10))
axs[0, 0].pie(counts)
axs[0, 0].legend(counts.index.values)
axs[0, 0].set_title('Most common car brands')

axs[0, 1].hist(veh.loc[veh["mileage(km)"] <= 1000000]["mileage(km)"], bins=1000)
axs[0, 1].set_title('Mileage(km)')

axs[1, 0].bar(["0-5 years", "5-10 years", "10-15 years", "15-20 years", "20-25 years"], [len(u5), len(fiveto10), len(tento15), len(fifteento20), len(twentyto25)])
axs[1, 0].set_title('Age distribution')

axs[1, 1].boxplot([gas, die])
axs[1, 1].set_xticklabels(['Gasoline', 'Diesel'])
axs[1, 1].set_title('CO2 emissions.')

fig.savefig("vehstats.png")
```
